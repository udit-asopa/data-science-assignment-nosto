```python
import ast
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

pd.set_option("display.max_columns", 100)
# (I always do this, otherwise pandas output is unreadable)
```

```python
# load data
visits = pd.read_csv("data/visits.tsv", sep="\t")
print("Shape:", visits.shape)
print(visits.dtypes)
visits.head()
# quick check: any missing values?
print("Missing values per column:\n", visits.isnull().sum())
```

```python
# parse product columns (viewed, bought, carted) from string to list, deduplicate, count
def parse_list_cell(x):
    """Convert cell from object/str to list[int]. Example: '[1, 2, 2]' -> [1, 2, 2]."""
    if isinstance(x, list):
        return [int(i) for i in x]
    if isinstance(x, str):
        try:
            parsed = ast.literal_eval(x)
            return [int(i) for i in parsed] if isinstance(parsed, list) else []
        except (ValueError, SyntaxError, TypeError):
            return []
    return []

def process_product_columns(
    df: pd.DataFrame,
    columns: list[str],
    deduplicate_keep_order: bool = True,
    add_count: bool = True,
    count_suffix: str = "total_",
) -> pd.DataFrame:
    out = df.copy()
    for col in columns:
        out[col] = out[col].apply(parse_list_cell)
        if deduplicate_keep_order:
            out[col] = out[col].apply(lambda xs: list(dict.fromkeys(xs)))
        if add_count:
            out[f"{count_suffix}{col}"] = out[col].apply(len)
    return out

product_columns = ["viewed_products", "bought_products", "put_in_cart_products"]
visits = process_product_columns(visits, product_columns)
print(visits.dtypes)
visits.head()
# sanity: check for any negative or weird counts
for col in [f"total_{c}" for c in product_columns]:
    print(col, ": min=", visits[col].min(), "max=", visits[col].max())
```

```python
# add datetime columns (start/end), hour, dayofweek
visits["end_dt"] = pd.to_datetime(visits["end"], unit="ms", utc=False)
visits["end_dt"] = visits["end_dt"].dt.floor("10ms")
visits["end_hour"] = visits["end_dt"].dt.hour
visits["end_dayofweek"] = visits["end_dt"].dt.dayofweek + 1  # 1=Mon
visits["end_date"] = visits["end_dt"].dt.date

# start_dt = end_dt - time_spent
visits["start_dt"] = visits["end_dt"] - pd.to_timedelta(visits["time_spent_in_minutes"], unit="m")
visits["start_dt"] = visits["start_dt"].dt.floor("10ms")
visits["start_hour"] = visits["start_dt"].dt.hour
visits["start_dayofweek"] = visits["start_dt"].dt.dayofweek + 1 # 1=Mon
visits["start_date"] = visits["start_dt"].dt.date

print(visits.dtypes)
visits.head()
# quick check: any negative time_spent?
print("Negative time_spent rows:", (visits["time_spent_in_minutes"] < 0).sum())
```

# Step 1 — Define the Target Variable

Goal: predict how long (in hours) until a customer returns after a visit.

Approach:
- Sort each customer's visits by start time
- For each visit, compute: `return_hours = (next_visit.start_dt − current_visit.end_dt)` in hours
- Last visit per customer has no next visit (so no target, drop those)
- This is a regression problem, target is right-skewed (lots of short gaps, some very long)

```python
# sort by customer, start_dt
visits_sorted = visits.sort_values(["customer_id", "start_dt"]).reset_index(drop=True)

# for each visit, get the start of the next visit for that customer
visits_sorted["next_start_dt"] = visits_sorted.groupby("customer_id")["start_dt"].shift(-1)

# target: hours from current visit end to next visit start
raw_return_hours = (
    visits_sorted["next_start_dt"] - visits_sorted["end_dt"]
).dt.total_seconds() / 3600

# quick sanity check - are there customers whose next visit starts before the current one ends? shouldn't happen but worth checking
overlap = visits_sorted[raw_return_hours < 0]
print(f"{len(overlap)} overlapping sessions - likely multi-device, clipping to 0")

# Negative gaps arise from overlapping sessions (e.g. multi-device) — treat as 0
n_negative = (raw_return_hours < 0).sum()
visits_sorted["return_hours"] = raw_return_hours.clip(lower=0)

# flag last visit per customer (no next visit ⇒ no target)
visits_sorted["is_last_visit"] = visits_sorted["next_start_dt"].isna()

# per-visit sequence index (0 = first ever visit for that customer)
visits_sorted["visit_index"] = visits_sorted.groupby("customer_id").cumcount()

# did the customer buy anything this visit?
visits_sorted["visit_bought_flag"] = visits_sorted["total_bought_products"] > 0

# summary
n_total       = len(visits_sorted)
n_with_target = visits_sorted["return_hours"].notna().sum()
n_last        = visits_sorted["is_last_visit"].sum()
n_unique_cust = visits_sorted["customer_id"].nunique()

print(f"Total rows                : {n_total:,}")
print(f"Unique customers          : {n_unique_cust:,}")
print(f"Rows with target          : {n_with_target:,}  ← used for modelling")
print(f"Last-visit rows (dropped) : {n_last:,}")
print(f"\nreturn_hours statistics (after clipping):")
print(visits_sorted["return_hours"].describe().round(1))
```

## Step 3 — Feature Engineering

features fall into a few buckets:
- what happened this session (views, buys, time spent)
- how long since last visit + rolling history
- cumulative purchase behaviour (leak-safe: shift before cumsum)

```python
# feature engineering (leak-safe)
# NOTE: originally did cummax() on current visit — that's leakage!
# must shift(1) first so we only look at *past* visits
fe = visits_sorted.copy()
# within-session signals
fe["buy_ratio"]  = np.where(fe["total_viewed_products"] > 0, fe["total_bought_products"] / fe["total_viewed_products"], 0.0)
fe["cart_ratio"] = np.where(fe["total_viewed_products"] > 0, fe["total_put_in_cart_products"] / fe["total_viewed_products"], 0.0)
# visit sequence
fe["past_visits_count"] = fe["visit_index"]   # alias for clarity
# ever_bought: has the customer EVER bought before this visit?
fe["ever_bought"] = (
    fe.groupby("customer_id")["visit_bought_flag"]
    .apply(lambda s: s.shift(1).fillna(False).cummax())
    .reset_index(level=0, drop=True)
    .astype(int)
)
# cum_bought_visits: how many previous visits resulted in a purchase?
fe["cum_bought_visits"] = (
    fe.groupby("customer_id")["visit_bought_flag"]
    .apply(lambda s: s.shift(1).fillna(False).cumsum())
    .reset_index(level=0, drop=True)
    .astype(int)
)
fe["cum_buy_rate"] = np.where(
    fe["past_visits_count"] > 0,
    fe["cum_bought_visits"] / fe["past_visits_count"],
    np.nan,      # undefined for first visit
)
# lag features: previous session values
fe["prev_end_dt"]      = fe.groupby("customer_id")["end_dt"].shift(1)
fe["prev_gap_hours"]   = (
    (fe["start_dt"] - fe["prev_end_dt"]).dt.total_seconds() / 3600
).clip(lower=0)
# NaN for first visit; non-negative (sessions shouldn't overlap)
fe["prev_time_spent"]   = fe.groupby("customer_id")["time_spent_in_minutes"].shift(1)
fe["prev_search_count"] = fe.groupby("customer_id")["num_of_times_search_was_used"].shift(1)
fe["prev_viewed_count"] = fe.groupby("customer_id")["total_viewed_products"].shift(1)
fe["prev_bought"]       = fe.groupby("customer_id")["visit_bought_flag"].shift(1).astype(float)
# rolling averages over last 3 past visits
def past_rolling_mean(series, window=3):
    return series.shift(1).rolling(window, min_periods=1).mean()
fe["rolling_avg_gap"]        = fe.groupby("customer_id")["prev_gap_hours"].transform(lambda s: past_rolling_mean(s))
fe["rolling_avg_time_spent"] = fe.groupby("customer_id")["time_spent_in_minutes"].transform(lambda s: past_rolling_mean(s))
fe["rolling_avg_viewed"]     = fe.groupby("customer_id")["total_viewed_products"].transform(lambda s: past_rolling_mean(s))
# customer-level expanding average gap (only past visits)
fe["cum_avg_gap"] = fe.groupby("customer_id")["prev_gap_hours"].transform(lambda s: s.shift(1).expanding().mean())
print("Feature engineering complete. fe.shape:", fe.shape)
# Show a sample of a repeat customer to verify no leakage
repeat_custsample = (
    fe.groupby("customer_id")["visit_index"].max()
    .pipe(lambda s: s[s >= 3]).index[0]
)
demo_cols = [
    "visit_index", "start_dt", "end_dt",
    "visit_bought_flag", "ever_bought", "cum_bought_visits", "cum_buy_rate",
    "prev_gap_hours", "cum_avg_gap", "return_hours",
]
print(f"\nSample rows for customer {repeat_custsample}:")
print(fe[fe["customer_id"] == repeat_custsample][demo_cols].to_string(index=False))
```

## Step 4 — Modeling

try HGB - handles NaNs natively so no need to impute first-visit rows

```python
FEATURE_COLS = [
    # within-session
    "total_viewed_products",
    "total_put_in_cart_products",
    "total_bought_products",
    "num_of_times_search_was_used",
    "time_spent_in_minutes",
    "buy_ratio",
    "cart_ratio",
    "start_hour",
    "start_dayofweek",
    "visit_bought_flag",
    # sequence
    "visit_index",
    # purchase history
    "ever_bought",
    "cum_bought_visits",
    "cum_buy_rate",
    # lag / recency
    "prev_gap_hours",
    "prev_time_spent",
    "prev_search_count",
    "prev_viewed_count",
    "prev_bought",
    # rolling averages
    "rolling_avg_gap",
    "rolling_avg_time_spent",
    "rolling_avg_viewed",
    # customer-level
    "cum_avg_gap",
]
# drop rows with no target (last visit per customer)
model_df = fe[fe["return_hours"].notna()].copy()
# chronological 80/20 split: train on earlier visits, validate on later
split_time = model_df["end_dt"].quantile(0.8)
train_df = model_df[model_df["end_dt"] <= split_time].copy()
val_df   = model_df[model_df["end_dt"]  > split_time].copy()
X_train = train_df[FEATURE_COLS]
X_val   = val_df[FEATURE_COLS]
y_train = train_df["return_hours"]
y_val   = val_df["return_hours"]
print(f"Features     : {len(FEATURE_COLS)}")
print(f"Model rows   : {len(model_df):,}  (excluded {len(fe) - len(model_df):,} last-visit rows)")
print(f"Split time   : {split_time}")
print(f"Train rows   : {len(train_df):,}")
print(f"Val rows     : {len(val_df):,}")
print(f"\ny_train stats:"); print(y_train.describe().round(1))
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

def mae(y_true, y_pred):
    return mean_absolute_error(y_true, y_pred)

def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))
# baseline: global median
median_global = y_train.median()
baseline_preds = np.full(len(y_val), median_global)
print("── Global Median Baseline ──────────────────────────────────────")
print(f"  Prediction value : {median_global:.1f}h  ({median_global/24:.1f} days)")
print(f"  MAE              : {mae(y_val, baseline_preds):.2f}h")
print(f"  RMSE             : {rmse(y_val, baseline_preds):.2f}h")
# segment-aware baseline: median per (ever_bought × start_dayofweek)
seg_medians = (
    train_df.groupby(["ever_bought", "start_dayofweek"])["return_hours"]
    .median()
    .rename("seg_median")
)
val_seg = val_df[["ever_bought", "start_dayofweek"]].join(
    seg_medians, on=["ever_bought", "start_dayofweek"]
)
val_seg["seg_median"] = val_seg["seg_median"].fillna(median_global)
seg_preds = val_seg["seg_median"].values
print("\n── Segmented Median Baseline (ever_bought × day-of-week) ──────")
print(f"  MAE  : {mae(y_val, seg_preds):.2f}h")
print(f"  RMSE : {rmse(y_val, seg_preds):.2f}h")
```

```python
# try HGB - handles NaNs natively so no need to impute first-visit rows
from sklearn.ensemble import HistGradientBoostingRegressor
y_train_log = np.log1p(y_train)
hgb = HistGradientBoostingRegressor(
    loss="absolute_error",    # optimises MAE directly
    learning_rate=0.05,
    max_iter=400,
    max_depth=6,
    min_samples_leaf=30,
    random_state=42,
)
hgb.fit(X_train, y_train_log)
# Predict and invert log transform; clip to 0 (no negative return times)
preds_log = hgb.predict(X_val)
preds_hgb = np.expm1(preds_log).clip(min=0)
print("── HistGradientBoosting (log1p target, MAE loss) ───────────────")
print(f"  MAE  : {mae(y_val, preds_hgb):.2f}h")
print(f"  RMSE : {rmse(y_val, preds_hgb):.2f}h")
# hm, baseline is winning - let me check if the val period just has a different distribution than train
## (see next cell for drift check)
print("\n── Summary comparison ──────────────────────────────────────────")
print(f"  {'Model':<35} {'MAE':>8} {'RMSE':>8}")
print(f"  {'-'*51}")
print(f"  {'Global Median Baseline':<35} {mae(y_val, baseline_preds):>8.2f} {rmse(y_val, baseline_preds):>8.2f}")
print(f"  {'Segmented Median (ever_bought × dow)':<35} {mae(y_val, seg_preds):>8.2f} {rmse(y_val, seg_preds):>8.2f}")
print(f"  {'HGB (23 features, log1p/MAE loss)':<35} {mae(y_val, preds_hgb):>8.2f} {rmse(y_val, preds_hgb):>8.2f}")
# quick scatter
plt.scatter(y_val, preds_hgb, alpha=0.1)
plt.xlabel("Actual return_hours"); plt.ylabel("Predicted")
plt.title("Actual vs Predicted (HGB)")
plt.show()
```

```python
# drift check: is the val period just different?
train_median = y_train.median()
val_median   = y_val.median()
drift_pct    = (abs(val_median - train_median) / train_median )* 100
print("── Target distribution: train vs val ───────────────────────────")
print(f"  Train median : {train_median:.1f}h  ({train_median/24:.1f}d)")
print(f"  Val   median : {val_median:.1f}h  ({val_median/24:.1f}d)")
print(f"  Drift        : {drift_pct:.1f}%")
if drift_pct > 10:
    print("  ⚠  Significant target drift detected — the val period has a")
    print("     different return-time distribution than train. This alone")
    print("     can widen MAE independently of model quality.")
else:
    print("  ✓  Target distributions are similar; drift is not a major factor.")
print(f"\n  Train date range: {train_df['start_dt'].min().date()} → {train_df['start_dt'].max().date()}")
print(f"  Val   date range: {val_df['start_dt'].min().date()} → {val_df['start_dt'].max().date()}")
```

```python
# feature importances (permutation-based)
from sklearn.inspection import permutation_importance
perm = permutation_importance(
    hgb, X_val, np.log1p(y_val),   # same log1p scale the model was trained on
    n_repeats=10,
    random_state=42,
    n_jobs=-1,
)
importances = (
    pd.Series(perm.importances_mean, index=FEATURE_COLS)
    .sort_values(ascending=True)
)
importances.plot.barh(figsize=(8, 7), color="steelblue", edgecolor="k", linewidth=0.4)
plt.axvline(0, color="k", linewidth=0.8, linestyle="--")
plt.xlabel("Mean decrease in MAE (log1p scale)")
plt.title("Permutation importances (HGB)")
plt.tight_layout()
plt.show()
print("\nTop 10 most important features:")
print(importances.sort_values(ascending=False).head(10).to_string())
```

## summary & what actually matters

- return time is super noisy and right-skewed; log1p helps but doesn't make it easy
- most customers are one-timers, so the train set is small after dropping last visits
- buyers do come back a bit sooner, but not by much; search usage doesn't really separate them
- the baseline (global median) is hard to beat — and in this case, it wins. that's not a failure, it's just what the data gives you
- the only feature that really matters is prev_gap_hours (how long since last visit). basically, if you haven't visited in a while, you probably won't anytime soon. everything else is minor.
- the validation period has a much lower median return time than train (target drift), so even a perfect model would struggle to "beat" the baseline if it's trained on the wrong distribution.

### what I'd try next with more time:
- survival model (Cox PH) to properly handle the last-visit censoring
- full year of data - 2 months isn't enough to capture seasonality
- reframe as classification: "returns within 7 days?" is more useful for marketing
